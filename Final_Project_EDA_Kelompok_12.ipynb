Dengan menggunakan gabungan beberapa syntax, preprocessing data teks pada kolom **data["Tweet"]** dapat dilakukan sebagai berikut.

```python
stopId, lemmaId = tau.LoadStopWords(lang='id')
slangFixId = tau.loadCorpus(file = 'data/slang.txt', sep=':')

cleaned_data = []
for tweet in tqdm(data['Tweet']):
    cleaned_data.append(tau.cleanText(tweet, lemma=lemmaId, lan='id', stops = stopId, fix=slangFixId))

cleaned_data[:7]
```

Terlihat bahwa pada data aslinya, kata "RT" sering muncul pada tweet-tweet tersebut. Oleh karena itu, kata tersebut ("rt") akan dimasukkan ke dalam stopwords sehingga pada data hasil preprocessingnya tidak terdapat lagi kata-kata tersebut.

```python
# stopId berupa set
stopId, lemmaId = tau.LoadStopWords(lang='id')
stopId.add("rt")
```

```python
"rt" in stopId
```

```python
cleaned_data = []
for tweet in tqdm(data["Tweet"]):
    cleaned_data.append(tau.cleanText(tweet, lemma=lemmaId, lan='id', stops = stopId, fix=slangFixId))

cleaned_data[:7]
```

Data setelah preprocessing:
- Tidak terdapat emoji, link, dan punctuation
- Semua data menjadi lowercase
- Beberapa kata telah mengalami lemmatization
- Beberapa slang telah ditangani
- Beberapa stopword telah terfilter

```python
data["Tweet Preprocessed"] = cleaned_data
```
